```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from bs4 import BeautifulSoup as bs
```

```python
data=requests.get("https://www.flipkart.com/search?q=shoes+men&sid=osp%2Ccil%2Ce1f&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_3_5_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_3_5_na_na_na&as-pos=3&as-type=RECENT&suggestionId=shoes+men%7CCasual+Shoes&requestId=59042497-fae5-4ab1-ae32-7265ed6204c6&as-backfill=on")
```

```python
soup=bs(data.content,'html.parser')
```

```python
soup
```

```python
shoes_info=soup.find_all('div',class_='_1xHGtK _373qXS')
```

```python
print(len(shoes_info))
```

```python
shoes=[]
for shoe in shoes_info:
    
    shoe_price = shoe.find('div', class_='_30jeq3')
    price = shoe_price
    com_name=shoe.find('div',class_='_2WkVRV')
    company=com_name.contents[-1].text
    shoe_dis=shoe.find('div',class_='_3Ay6Sb')
    discount=shoe_dis.contents[-1].text.split()[0]
    
    shoes_dict={'price':price,
                'company':company,
                'discount':discount}
    shoes.append(shoes_dict)
```

```python
shoes
```

```python
shoes_df=pd.DataFrame(shoes)
```
